<h2>4. Package Management & Dependencies</h2>

<h3>pip (Python Package Installer)</h3>
<p><strong>Standard package manager for Python</strong>

<p><strong>Key Features:</strong>
<ul>
<li>Install packages from PyPI</li>
<li>Manage dependencies</li>
<li>Virtual environment support</li>
<li>Requirements.txt support</li>
</ul>

<p><strong>Essential Commands:</strong>
```bash
pip install package_name     # Install package
pip install -r requirements.txt  # Install from file
pip freeze > requirements.txt    # Generate requirements
pip list                      # List installed packages
```

<h3>conda (Cross-Platform Package Manager)</h3>
<p><strong>Package and environment management system</strong>

<p><strong>Key Features:</strong>
<ul>
<li>Cross-platform package management</li>
<li>Environment isolation</li>
<li>Binary package distribution</li>
<li>Dependency resolution</li>
</ul>

<p><strong>Essential Commands:</strong>
```bash
conda create -n env_name python=3.11  # Create environment
conda activate env_name                # Activate environment
conda install package_name            # Install package
conda env export > environment.yml    # Export environment
```

<h3>poetry (Modern Dependency Management)</h3>
<p><strong>Dependency management and packaging tool</strong>

<p><strong>Key Features:</strong>
<ul>
<li>Dependency resolution</li>
<li>Virtual environment management</li>
<li>Build and publish packages</li>
<li>Lock file for reproducible builds</li>
</ul>

<h3>uv (Ultra-Fast Package Installer)</h3>
<p><strong>Next-generation Python package installer</strong>

<p><strong>Key Features:</strong>
<ul>
<li>Extremely fast installation</li>

<h2>5. Code Quality & Testing Tools</h2>

<h3>Black (Code Formatter)</h3>
<p><strong>Uncompromising Python code formatter</strong>

<p><strong>Key Features:</strong>
<ul>
<li>Automatic code formatting</li>
<li>Consistent code style</li>
<li>Integration with IDEs</li>
<li>Pre-commit hooks support</li>
</ul>

<p><strong>Usage:</strong>
```bash
black your_file.py              # Format single file
black .                         # Format entire project
black --check .                 # Check formatting without changes
```

<h3>flake8 (Linting Tool)</h3>
<p><strong>Style guide enforcement and error detection</strong>

<p><strong>Key Features:</strong>
<ul>
<li>PEP 8 style checking</li>
<li>Syntax error detection</li>
<li>Unused imports detection</li>
<li>Complexity analysis</li>
</ul>

<h3>pytest (Testing Framework)</h3>
<p><strong>Modern Python testing framework</strong>

<p><strong>Key Features:</strong>
<ul>
<li>Simple test discovery</li>
<li>Powerful fixtures</li>
<li>Parametrized testing</li>
<li>Rich plugin ecosystem</li>
</ul>

<p><strong>Example Test:</strong>
```python
def test_add_numbers():
    assert add(2, 3) == 5
    assert add(-1, 1) == 0
```

<h3>mypy (Static Type Checking)</h3>
<p><strong>Optional static typing for Python</strong>

<p><strong>Key Features:</strong>
<ul>
<li>Type checking without runtime overhead</li>
<li>Better IDE support</li>
<li>Catches type-related bugs</li>
<li>Gradual adoption</li>
</ul>
